```python
from sklearn.linear_model import LinearRegression
model0 = LinearRegression(normalize=True)
ximgs = np.array(OnlineNews[' num_imgs'])
y = np.array(OnlineNews[' shares'])
Ximgs = ximgs[:, np.newaxis]
print(y)
model0.fit(Ximgs, y) 
print(model0.coef_)
print(model0.intercept_)
plt.plot(ximgs, y, 'o')
plt.xlim(0, 40)
plt.ylim(0, 300000)
```

Looking at the relationship bewteen number of images and shares, there is not a great correlation. The number images appears to help increase the shares to a point. Too many images is clearly no help towards the popularity of an article. A linear model might not be appropriate here. 

There are a significant amount more features in the dataset that can be used to fit a model, as oppossed to just images. The dataset also includes a number of categorial features, which could really help fit the right model.
